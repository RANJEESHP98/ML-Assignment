```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = load_breast_cancer()
df=pd.DataFrame(data.data,columns=data.feature_names)
df
df['target_value'] = data.target
df.head()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(corr_matrix,vmin=-0.7,vmax=0.7,annot=True,linewidths=0.2,cmap='YlGnBu')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# split the datasets into x and y
x = df.drop(columns=['target_value'])
y= df['target_value']
```

```python
data.target_names
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state=42)
```

```python
# for scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Create logistic regression model
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
y_predict = log_model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score, accuracy_score
```

```python
x_test.shape
```

```python
print('Based on the evaluation:')
print('Accuracy score ',accuracy_score(y_test,y_predict))
print('Precision score',precision_score(y_test,y_predict))
print('Recall score',recall_score(y_test,y_predict))
print('F1 score', f1_score(y_test,y_predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict,target_names=data.target_names))
```

```python
# confusion metrix
from sklearn.metrics import confusion_matrix
log_confusion = confusion_matrix(y_test,y_predict)
print('confusion matrix',log_confusion)
```

```python
# Visulization uisng Heatmap of confusion matrix
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.heatmap(log_confusion,annot=True,cmap='Blues',fmt ='d',xticklabels=['malignant','benign'],yticklabels=['malignant','benign'])
plt.title('Confusion Matrix')
plt.show()
```

 Based on findings : based on the actual and predictions the 96% the  True possitive predictions are accurate.Meanwhile,comparing weighted average 
and macro average is slightly difference in precision model others are almost equall.

Comparison

```python
# splitting the weighted and macro for comparing the models.
#macro average
macro_avg_lg = f1_score(y_test,y_predict,average='macro')
macro_avg_dt = f1_score(y_test,y_predict_dt,average='macro')
macro_avg_rfc = f1_score(y_test,y_predict_rfc,average='macro')
macro_avg_svc = f1_score(y_test,y_predict_svc,average='macro')
macro_avg_knn = f1_score(y_test,y_predict_knn,average='macro')
```

```python
# weighted average
weighted_avg_lg = f1_score(y_test,y_predict,average='weighted')
weighted_avg_dt = f1_score(y_test,y_predict_dt,average='weighted')
weighted_avg_rfc = f1_score(y_test,y_predict_rfc,average='weighted')
weighted_avg_svc = f1_score(y_test,y_predict_svc,average='weighted')
weighted_avg_knn = f1_score(y_test,y_predict_knn,average='weighted')
```

```python
# baesed on the f1score compare the weighted average and macro average in that choose the worst model and best model.
cm_model = {
    'Model': ['Logistic Regression', 'Decision Tree', 'Random Forest Classifier', 
              'Support Vector Machine', 'K-Nearest Neighbours'],
    'Weighted_average': [weighted_avg_lg, weighted_avg_dt, weighted_avg_rfc, 
                         weighted_avg_svc, weighted_avg_knn],
    'macro_average':[macro_avg_lg,macro_avg_dt,macro_avg_rfc,macro_avg_svc,macro_avg_knn]
}

fl_model = pd.DataFrame(cm_model)
print(fl_model)

```

Findings are based on the f1score the random forest classifier is best model for performing and finding the weighted and macro,
secondly, the Logistic regression model , it really works with linearly seperable class.Thirdly,Knn model classified in moderate way 
for scaling the data. Lastly, Decision tree and SVM(support sector machine) classified in least effective way.
